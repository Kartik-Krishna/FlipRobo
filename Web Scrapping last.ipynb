# Scrape the details of most viewed videos on YouTube from Wikipedia

```python
url1 = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url1)
time.sleep(3)                                         #accessing the url
```

```python
rank = []
name = []
artist = []
up_date = []
views = []                                                           #declaring lists

top_vid = driver.find_element_by_xpath('//li[@class="toclevel-1 tocsection-2"]//a')
top_vid.click()
time.sleep(1)

for i in range (1, 31):                                              #using for loop to loop over all the rows at once
    xp = '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr['
    rank_path = driver.find_element_by_xpath(xp + str(i) + ']//td[1]')
    name_path = driver.find_element_by_xpath(xp + str(i) + ']//td[2]')
    artist_path = driver.find_element_by_xpath(xp + str(i) + ']//td[3]')
    up_date_path = driver.find_element_by_xpath(xp + str(i) + ']//td[4]')
    views_path = driver.find_element_by_xpath(xp + str(i) + ']//td[5]')
    
    rank.append(rank_path.text)
    name.append(name_path.text)
    artist.append(artist_path.text)
    up_date.append(up_date_path.text)
    views.append(views_path.text)
    
wiki = pd.DataFrame({})                                             #creating dataframe
wiki['Rank'] = rank
wiki['Name'] = name
wiki['Artist'] = artist
wiki['Upload Date'] = up_date
wiki['Views'] = views
wiki.set_index('Rank', inplace = True)
wiki
```

```python
wiki['Name'] = wiki['Name'].apply(lambda x: x[1:-5])        #removing quotation marks and [i]
wiki
```

# Scrape the details team India’s international fixtures from bcci.tv.

```python
url2 = 'https://www.bcci.tv/'
driver.get(url2)
time.sleep(3)                                             #accessing the url

intnat = driver.find_element_by_xpath('//li[@data-nav-index="0"]//div[2]')
intnat.click()
time.sleep(1)

fix = driver.find_element_by_xpath('//li[@data-nav-index="0"]//div[2]//li[1]//a')
fix.click()
time.sleep(3)
```

```python
title = []
series = []
place = []
date = []
ti = []                                                  #declaring lists

title_tag = driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
series_tag = driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
place_tag = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//span')
date_tag = driver.find_elements_by_xpath('//div[@class="fixture__full-date"]')
time_tag = driver.find_elements_by_xpath('//span[@class="fixture__time"]')

for i in title_tag:                                                  #looping over tag and adding relevant text to the list
    title.append(i.text)
    
for i in series_tag:                                                #looping over tag and adding relevant text to the list
    series.append(i.text)
    
for i in place_tag:                                                 #looping over tag and adding relevant text to the list
    place.append(i.text)
    
for i in date_tag:                                                 #looping over tag and adding relevant text to the list
    date.append(i.text.replace('\n', ' '))
    
for i in time_tag:                                                #looping over tag and adding relevant text to the list
    ti.append(i.text)
    
bcci = pd.DataFrame({})                                               #creating dataframe
bcci['Title'] = title
bcci['Series'] = series
bcci['Place'] = place
bcci['Date'] = date
bcci['Time'] = ti
bcci
```

```python
bcci['Date'] = bcci['Date'].apply(lambda x: x[:-10])          #removing time from date column
bcci
```

# Scrape the details of State-wise GDP of India from statisticstime.com.

```python
url4 = 'http://statisticstimes.com/'                     #accessing the url
driver.get(url4)
time.sleep(3)

eco = driver.find_element_by_xpath('//div[@class="dropdown"][2]')
hover = ActionChains(driver).move_to_element(eco)
hover.perform()                             #this will hover the mouse cursor over the element so drop down menu will appear

ind = driver.find_element_by_xpath('//div[@class="dropdown"][2]//a[3]')
ind.click()
time.sleep(2)                                #now we are able to access dropdown menu options

gdp_page = driver.find_element_by_xpath('//ul[@style="list-style-type:none;margin-left:20px;"]//li[1]')
gdp_page.click()
time.sleep(2)
```

```python
rank = []
name = []
gsdp_18_19 = []
data = []                         #declaring lists

rank_tag = driver.find_elements_by_xpath('//td[@class="data1"]')
data_tag = driver.find_elements_by_xpath('//td[@class="data"]')
gsdp_tag = driver.find_elements_by_xpath('//td[@class="data sorting_1"]')
name_tag = driver.find_elements_by_xpath('//td[@class="name"]')

for i in rank_tag:                                    #looping over tag and adding relevant text to the list
    rank.append(i.text)

for i in name_tag:                                    #looping over tag and adding relevant text to the list
    name.append(i.text)
    
for i in data_tag:                                    #looping over tag and adding relevant text to the list
    data.append(i.text)

for i in gsdp_tag:                                  #looping over tag and adding relevant text to the list
    gsdp_18_19.append(i.text)

economy = pd.DataFrame({})                          #creating dataframe
economy['Rank'] = rank[:33]                         #will access till 33rd element of list because beyond that 
economy['State'] = name[:33]                        #other table's content will start which are not relavent to us
economy['GSDP 19-20'] = data[0:161:5]
economy['GSDP 18-19'] = gsdp_18_19[:33] 
economy['Share'] = data[1:162:5]
economy['GDP($billion)'] = data[2:163:5]
economy
```

# Scrape the details of trending repositories on Github.com.

```python
url5 = 'https://github.com/'
driver.get(url5)
time.sleep(3)                            #accessing the url

exp = driver.find_element_by_xpath('//li[@class="d-block d-lg-flex flex-lg-nowrap flex-lg-items-center border-bottom border-lg-bottom-0 mr-0 mr-lg-3 edge-item-fix position-relative flex-wrap flex-justify-between d-flex flex-items-center "][2]//summary')
hover = ActionChains(driver).move_to_element(exp)
hover.perform()                           #this will hover the mouse cursor over the element so drop down menu will appear

trending = driver.find_element_by_xpath('//ul[@class="list-style-none mb-3"][2]//li[3]//a')
trending.click()
time.sleep(2)
```

```python
href = []
repo = []
desc = []
cont = []
lang = []                                 #declaring lists

repo_tag = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]//a')

for i in repo_tag:
    repo.append(i.text)
    href.append(i.get_attribute('href'))            #getting href of all the repos in 1 list
    
for i in href:
    driver.get(i)
    time.sleep(3)                                 #accessing all the repos 1 by 1
    
    try:
        a = 1
        while a < 7:
            cont_tag = driver.find_element_by_xpath('//div[@class="BorderGrid-row"]['+ str (a) + ']//a')
            if ('Contributors' in cont_tag.text):
                cont.append(cont_tag.text)
                break
    
            else:
                a = a+1
    
    except:
        cont.append('-')
    
    try:
        a = 1
        while a < 7:
            lang_tag = driver.find_element_by_xpath('//div[@class="BorderGrid-row"]['+ str (a) + ']')
            if ('Languages' in lang_tag.text):
                language_tag = driver.find_element_by_xpath('//div[@class="BorderGrid-row"]['+ str (a) + ']//ul')
                lang.append(language_tag.text.replace('\n', ' '))
                break
    
            else:
                a = a+1
        
    except:
        lang.append('-')
        
    try:
        desc_tag = driver.find_element_by_xpath('//div[@class="BorderGrid-row hide-sm hide-md"]//p')
        desc.append(desc_tag.text)
    except:
        desc_tag = driver.find_element_by_xpath('//div[@class="f4 mt-3 color-text-secondary text-italic"]')
        desc.append(desc_tag.text)
        
git = pd.DataFrame({})                           #creating dataframe
git['Repository'] = repo
git['Description'] = desc
git['Contributor Count'] = cont
git['Languages'] = lang

git
```

```python
git['Contributor Count'] = git['Contributor Count'].apply(lambda x: x[13:])
git
```

# Scrape the details of Data science recruiters from naukri.com.

```python
url7 = 'https://www.naukri.com/'
driver.get(url7)
time.sleep(3)                               #accessing the url

rec = driver.find_element_by_xpath('//a[@title="Search Recruiters"]')
rec.click()
time.sleep(2)                                 #on clicking the link

driver.switch_to.window(driver.window_handles[1])              #the page opens on new tab, this will move us to that tab

search = driver.find_element_by_xpath('//div//input[@class="sugInp"]')
search.clear()
search.send_keys('Data Science')

search_button = driver.find_element_by_id('qsbFormBtn')
search_button.click()                               #the search will be executed
```

```python
name = []
des = []
comp = []
skills = []
loc = []
href = []                                             #declaring lists

name_path = driver.find_elements_by_xpath('//span[@class="fl ellipsis"]')
des_path = driver.find_elements_by_xpath('//span[@class="ellipsis clr"]')
comp_path = driver.find_elements_by_xpath('//a[@class="ellipsis"]')
skills_path = driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
loc_path = driver.find_elements_by_xpath('//small[@class="ellipsis"]')

for i in name_path:                                  #looping over tag and adding relevant text to the list
    name.append(i.text)
    
for i in des_path:                                 #looping over tag and adding relevant text to the list
    des.append(i.text)
    
for i in comp_path:                                        #looping over tag and adding relevant text to the list
    comp.append(i.text) 
    href.append(i.get_attribute('href'))
    
for i in skills_path:                             #looping over tag and adding relevant text to the list
    skills.append(i.text)
        
for i in href[0::2]:
    driver.get(i)
    time.sleep(3)
    
    try:
        loc_tag = driver.find_element_by_xpath('//div[@class="oh"][2]//a')
        loc.append(loc_tag.text)
    except:
        loc.append('-')
        
naukri = pd.DataFrame({})                                      #creating dataframe
naukri['Name'] = name
naukri['Designation'] = des
naukri['Company'] = comp[::2]
naukri['Skills'] = skills
naukri['Location'] = loc
naukri
```

# Scrape the details of Highest selling novels.

```python
book = []
author = []
vol = []
pub = []
genre = []                                     #declaring lists

url8 = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/'
driver.get(url8)
time.sleep(3)                               #accessing the url

for i in range (0,100):                                   #using for loop to loop over all the rows at once
    book_tag = driver.find_element_by_id('table-cell-10943-' + str(i) + '-1')
    auth_tag = driver.find_element_by_id('table-cell-10943-' + str(i) + '-2')
    vol_tag = driver.find_element_by_id('table-cell-10943-' + str(i) + '-3')
    pub_tag = driver.find_element_by_id('table-cell-10943-' + str(i) + '-4')
    gen_tag = driver.find_element_by_id('table-cell-10943-' + str(i) + '-5')
    
    book.append(book_tag.text)
    author.append(auth_tag.text)
    vol.append(vol_tag.text)
    pub.append(pub_tag.text)
    genre.append(gen_tag.text)
    
novels = pd.DataFrame({})                            #creating dataframe
novels['Title'] = book
novels['Author'] = author
novels['Volumes Sold'] = vol
novels['Publisher'] = pub
novels['Genre'] = genre
novels
```

# Scrape the details most watched tv series of all time from imdb.com.

```python
url9 = 'https://www.imdb.com/list/ls095964455/'
driver.get(url9)
time.sleep(3)                                   #accessing the url

name = []
year = []
genre = []
run = []
ratings = []
votes = []                                          #declaring lists

name_tag = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//a')
year_tag = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
genre_tag = driver.find_elements_by_xpath('//span[@class="genre"]')
run_tag = driver.find_elements_by_xpath('//span[@class="runtime"]')
ratings_tag = driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]//span[2]')
votes_tag = driver.find_elements_by_xpath('//span[@name="nv"]')

for i in name_tag:                                             #looping over tag and adding relevant text to the list
    name.append(i.text)
    
for i in year_tag:                                              #looping over tag and adding relevant text to the list
    year.append(i.text)
    
for i in genre_tag:                                         #looping over tag and adding relevant text to the list
    genre.append(i.text)
    
for i in run_tag:                                              #looping over tag and adding relevant text to the list
    run.append(i.text)
    
for i in ratings_tag:                                         #looping over tag and adding relevant text to the list
    ratings.append(i.text)
    
for i in votes_tag:                                          #looping over tag and adding relevant text to the list
    votes.append(i.text)
    
imdb = pd.DataFrame({})                                              #creating dataframe
imdb['Name'] = name
imdb['Year Span'] = year
imdb['Genre'] = genre
imdb['Run time'] = run
imdb['Ratings'] = ratings
imdb['Votes'] = votes

imdb
```

# Details of Datasets from UCI machine learning repositories.

```python
url10 = 'https://archive.ics.uci.edu/'                 
driver.get(url10)
time.sleep(3)                                        #accessing the url

ds = driver.find_element_by_xpath('//font[@size="3"]')
ds.click()
time.sleep(2)
```

```python
ds_name = []
dtype = []
task = []
att = []
no_inst = []
no_att = []
year = []                                                #declaring lists

total_tag = driver.find_element_by_xpath('//table[@width="100%"]//p[@class="big"]//b')
total = int (total_tag.text) + 2

for i in range (2, total):          #creating a list of list in which each element will contain all the column info in a list
    
    data_tag = driver.find_elements_by_xpath('//table[@border="1"]//tr[' + str(i) + ']//p')
    ds_name.append(data_tag[0].text)
    dtype.append(data_tag[1].text)
    task.append(data_tag[2].text)
    att.append(data_tag[3].text)
    no_inst.append(data_tag[4].text)
    no_att.append(data_tag[5].text)
    year.append(data_tag[6].text)
    
uci = pd.DataFrame({})                                             #creating dataframe
uci['Dataset name'] = ds_name
uci['Data type'] = dtype
uci['Task'] = task
uci['Attribute type'] = att
uci['No of instances'] = no_inst
uci['No of attribute'] = no_att
uci['Year'] = year

uci
```
