# Write a python program to scrape the salary data for Data Scientist designation in Noida location.

```python
url5 = 'https://www.glassdoor.co.in/Salaries/index.htm'
driver.get(url5)
time.sleep(4)

search_job = driver.find_element_by_id('KeywordSearch')
search_job.clear()
search_job.send_keys('Data Scientist')

search_loc = driver.find_element_by_id('LocationSearch')
search_loc.clear()
search_loc.send_keys('Noida')

search_button = driver.find_element_by_id('HeroSearchButton')
search_button.click()
```

```python
title_tags = driver.find_elements_by_xpath('//p[@class ="m-0 "]')
min_sal_tags = driver.find_elements_by_xpath('//div[@class="common__RangeBarStyle__values d-flex justify-content-between "]//span[1]')
max_sal_tags = driver.find_elements_by_xpath('//div[@class="common__RangeBarStyle__values d-flex justify-content-between "]//span[2]')
avg_sal_tags = driver.find_elements_by_xpath('//div[@class="col-2 d-none d-md-flex flex-row justify-content-end"]//strong')

titles = []
min_sal = []
max_sal = []
avg_sal = []

for i in title_tags:
    titles.append(i.text)

for i in min_sal_tags:
    min_sal.append(i.text)
    
for i in max_sal_tags:
    max_sal.append(i.text)
    
for i in avg_sal_tags:
    avg_sal.append(i.text)
    
salary = pd.DataFrame({})
salary['Company'] = titles
salary['Average Salary'] = avg_sal
salary['Minimum Salary'] = min_sal
salary['Maximum Salary'] = max_sal
salary = salary[:10]
salary
```

# Scrape data of first 100 sunglasses listings on flipkart.com.

```python
url6 = 'https://www.flipkart.com/'
driver.get(url6)
```

```python
cross = driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
cross.click()
time.sleep(5)
```

```python
search = driver.find_element_by_xpath('//div[@class="_3OO5Xc"]//input')
search.clear()
search.send_keys('sunglasses')

search_button = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()
time.sleep(5)

brand = []
description = []
price = []
discount = []


for a in range (3):
    brand_tags = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    desc_tags = driver.find_elements_by_xpath('//div[@class="_2B099V"]//a')
    price_tags = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    discount_tags = driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]//span')

    for i in brand_tags:
        brand.append(i.text)
    
    for i in desc_tags:
        description.append(i.text)
    
    for i in price_tags:
        price.append(i.text)
    
    for i in discount_tags:
        discount.append(i.text)

    if (a < 1):
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
        nxt.click()
        time.sleep(5)
    else:
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')
        nxt.click()
        time.sleep(5)

desc = description[0::2]
```

```python
sunglasses = pd.DataFrame({})
sunglasses['Brand'] = brand
sunglasses['Description'] = desc
sunglasses['Price'] = price
sunglasses['Discount'] = discount
sunglasses = sunglasses[:100]
sunglasses
```

# Scrape 100 reviews data from flipkart.com for iphone11 phone.

```python
url7 = 'https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace'
driver.get(url7)
all_r = driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]')
all_r.click()
time.sleep(3)
```

```python
rate = []
rev_sum = []
rev_full = []

for a in range (10):
    rate_tags = driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]//div[@class="row"]')
    rev_sum_tags = driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    rev_full_tags = driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    read_more_tags = driver.find_elements_by_xpath('//span[@class="_1BWGvX"]')

    for i in rate_tags:
        rate.append(i.text)
    
    for i in rev_sum_tags:
        rev_sum.append(i.text)
    
    for i in read_more_tags:
        i.click()
    
    for i in rev_full_tags:
        rev_full.append(i.text.replace('\n', ' '))
    
    if (a < 1):
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
        nxt.click()
            
    else:
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')
        nxt.click()
            
    time.sleep(3)

ra = rate[0::4]
```

```python
review = pd.DataFrame({})
review['Ratings'] = ra
review['Review Summary'] = rev_sum
review['Full Review'] = rev_full
review
```

```python
review['Ratings'] = review['Ratings'].apply(lambda x: x[:1])
review
```

# Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” and “Intel Core i9” After setting the filters scrape first 10 laptops data.

```python
url10 = 'https://www.amazon.in/'
driver.get(url10)

time.sleep(3)

search_bar = driver.find_element_by_id('twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys('laptops')

search = driver.find_element_by_id('nav-search-submit-button')
search.click()
time.sleep(3)

filter1 = driver.find_element_by_xpath('//li[@id="p_n_feature_thirteen_browse-bin/12598163031"]//a')
filter1.click()
time.sleep(3)

href = []
title = []
ratings = []
price = []

get_href = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]//a')

for i in get_href:
    href.append(i.get_attribute('href'))
    
for i in href:
    driver.get(i)
    time.sleep(3)
    title_tags = driver.find_element_by_xpath('//h1[@id="title"]')
    title.append(title_tags.text)
    try:
        ratings_tags = driver.find_element_by_xpath('//td[@class="a-size-base"]')
        ratings.append(ratings_tags.text)
    except:
        ratings.append('-')
    try:
        price_tags = driver.find_element_by_xpath('//span[@id="priceblock_ourprice"]')
        price.append(price_tags.text)
        
    except:
        price.append('-')
        
print (title)
print (ratings)
print (price)
```

```python
laptop1 = pd.DataFrame({})
laptop1['Title'] = title
laptop1['Rating'] = ratings
laptop1['Price'] = price
laptop1
```

```python
laptop1['Rating'] = laptop1['Rating'].apply(lambda x: x[-18:])
laptop1['Rating']
```
